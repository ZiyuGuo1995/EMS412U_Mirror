## 1.1 Line plot

Now, we give an example of line plot. You can see some commonly used control sentences in the example.

```python
# import all necessary packages
import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(0, 10, 100)        # Create x value, from 0 to 10 and 100 points in total.
y = 4 + 2 * np.sin(2 * x)          # Create y value, using equation 4 + 2sin(2x)

# plot
fig, ax = plt.subplots()           # Create a plot object.

ax.plot(x, y, linewidth=2.0)       # Plot with x and y. Also control the line width in the plot

ax.set(xlim=(0, 10), xticks=np.arange(0, 10,1),  #Set range of x and y axis(xlim and ylim), and set ticks on x and y axis.
       ylim=(0, 8), yticks=np.arange(0, 10,1))

ax.set_xlabel('X')                 # Axis Title
ax.set_ylabel('Y')                 # Axis Title
ax.set_title('Line Plot')          # Plot title

# plt.grid()                       # Create grid on the plot
plt.show()
```

## 1.2 Bar chart

Here is an example of a bar chart. we can simply call ax.bar() to create the chart. Format control sentences, such as axis range set, label set, and tick set, are the same as the line graph.

```python
import matplotlib.pyplot as plt
import numpy as np

# make data:
np.random.seed(3)
x = 0.5 + np.arange(8)
y = np.random.uniform(2, 7, len(x))

# plot
fig, ax = plt.subplots()

ax.bar(x, y, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

ax.set_xlabel('X') # Axis Title
ax.set_ylabel('Y') # Axis Title

plt.show()
```

### 1.3 Scatter
The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis. You can read the following code as well as comments for a description of each control statement.

```python
import matplotlib.pyplot as plt
import numpy as np

# make data
x = np.linspace(0, 10, 50)        # Create x value, from 0 to 10 and 100 points in total.
y = 4 + 2 * np.sin(2 * x)          # Create y value, using equation 4 + 2sin(2x)

# plot
fig, ax = plt.subplots()           # Create a plot object.

ax.scatter(x, y, c='r', marker='o')# Scatter plot with x and y. 
# c: color control, 'r': red, 'g': green, 'b': blue, 'y':yellow. 
# marker: shape control, 'o': circle, '.': point, 'v' Triangle down, '^': Triangle up,'s': square

ax.set(xlim=(0, 10), xticks=np.arange(0, 10,1),  #Set range of x and y axis(xlim and ylim), and set ticks on x and y axis.
       ylim=(0, 8), yticks=np.arange(0, 10,1))

ax.set_xlabel('X')                 # Axis Title
ax.set_ylabel('Y')                 # Axis Title
ax.set_title('Line Plot')          # Plot title

# plt.grid()                       # Create grid on the plot
plt.show()
```
